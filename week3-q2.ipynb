First of all, import libraries

```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
```

Second, Import data

```python
link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

tables = pd.read_html(link, header=0)
df = tables[0]
```

Drop lines which "borough" is "Not assigned"

```python
df_Borough = df[df["Borough"] != "Not assigned"]
```

Reset index numbers

```python
df_Borough_index = df_Borough.reset_index()
```

```python
df_BNi = df_Borough_index.iloc[:,1:]
```

Shows rows and columns

```python
df_BNi.shape
```

```python
df_BNi
```

Read the Geocoder package.

```python
geo_data = pd.read_csv('Geospatial_Coordinates.csv')
geo_data.head()
```

Merge two tables based on Postal Code.

```python
df_geo = pd.merge(df_BNi, geo_data, on='Postal Code')
df_geo.head()
```

Make table which is shown.

```python
df_geo_sample = df_geo[df_geo["Postal Code"].isin(['M5G','M2H','M4B','M1J','M4G','M4M','M1R','M9V','M9L','M5V','M1B','M5A'])]
```

```python
d_order = {'M5G': 0, 'M2H': 1, 'M4B': 2, 'M1J': 3,'M4G':4,'M4M':5,'M1R':6,'M9V':7,'M9L':8,'M5V':9,'M1B':10,'M5A':11}

df_geo_sample['order'] = df_geo_sample['Postal Code'].map(d_order)
```

```python
df_geo_sample_sort = df_geo_sample.sort_values("order")
```

```python
df_geo_sample_sort2 = df_geo_sample_sort.reset_index()
df_Final = df_geo_sample_sort2.iloc[:,1:-1]
```
